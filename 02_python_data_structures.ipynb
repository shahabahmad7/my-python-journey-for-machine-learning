# Data structures in Python
 
 ### Lists

```python
# Access through indexing 
A = ["Shahab Ahmad", 23, 3.2]
print(A[1])

# Slicing 
print(A[0:2])

#Adding elemnents using extend() and append()
B = ["Shahab Ahmad", 23, 3.2]
B.extend(["saidu sharif", 46])
print(B)

B = ["Shahab Ahmad", 23, 3.2]
B.append(["saidu sharif", 46])
print(B)

#List are mutable i.e we can change elements in it.
B = ["Shahab Ahmad", 23, 3.2]
B[0] = "Aizaz Ahmad"
print("After change", B)

# Deleting element from the list
C = ["Shahab Ahmad", 23, 3.2, "Abbas"]
del(C[3])
print("After Deleting", C)

#copy list
C = ["Shahab Ahmad", 23, 3.2, "Abbas"]     #changes in the one list will be reflected in other list too
E = C 
print("Copied", E)
C[0] = "Hazrat Usman"
print("Replace in the other list too", E)

 #Cloning in Lists
C = ["Shahab Ahmad", 23, 3.2, "Abbas"]    #changes will not be reflected in the D.
D = C [:]
print("Cloned", D)
```
