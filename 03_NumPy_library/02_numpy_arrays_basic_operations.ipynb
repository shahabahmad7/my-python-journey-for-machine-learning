### 1 Dimension array

```python
# First import the NumPy library
import numpy as np
```

```python
a1 = np.array([20,40,50,100,130])    # this is called one dimension array 

# we can access the value through indexing 
print(a1[2])
```

```python
# we can checked the different properties of an array.
a1.dtype, a1.ndim, a1.shape, a1.size, type(a1)

```

#### Slicing

```python
a1 = np.array([20,40,50,100,130]) 
a = a1[0:3]
print(a)
```

```python
# we can also change elements by this method
a1[2] = 22
a1
```

```python
# simple statistical method 
mean = a1.mean()
print("mean", mean)

standard_deviation= a1.std()
print("standard deviation", standard_deviation)

# we can also find the min and max value using min() and max().
max_value = a1.max()
min_value = a1.min()
print("max", max_value)
print("min", min_value)
```

Addition and Subtraction

```python
#  addition of an array 
arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])
add_arr = np.add(arr1, arr2)
print(add_arr)
```

```python
# Subtraction of an array
arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])
sub_arr = np.subtract(arr1, arr2)
print(sub_arr)
```

Multiplication and Division

```python
# we will do the multiplication on the same array 
arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])
multip_arr = np.multiply(arr1, arr2)
print(multip_arr)
 
 # division 
arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])
div_arr = np.divide(arr1, arr2)
print(div_arr)
```

#### Dot Product
dot product is different from the above traditional multiplication, in this the value are first multiplied and then added to give us value.

```python
x = np.array([3, 5])
y = np.array([2, 4])
z = np.dot(x, y)
print(z)

```
