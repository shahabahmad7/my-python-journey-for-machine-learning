## Functions in Python

```python
# a simple function 
def add(a, b):
    c = a + b
    return c
x = add(10,15)
print(x)


def greet(name):
    return f"hello, {name}"
greeting = greet("shahab")
print(greeting)
```

```python
def introduce (name, age, city):
    return f"{name} who is {age} years old, and he lives in the {city}."
introduction = introduce("Shahab", 23, 'Saidu Sharif')
print(introduction)
```
