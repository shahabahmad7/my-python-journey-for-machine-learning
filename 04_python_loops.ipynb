## Loops
 For Loop
  

```python
colors = ["red", "orange", "yellow", "green", "blue", "indigo", "violet"]
for color in colors: 
    print(f"For rainbow, one of the colour is {color}")
```

```python
# using enumerate to get index and value of list items
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
for fruit in enumerate(fruits):
    print(f"I like to eat {fruit}")

    # Looping through a range of numbers  upto 10  
for i in range(1, 11):
    print(f"The current number is {i}")
```
