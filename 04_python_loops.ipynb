## Loops
 For Loop
  

```python
colors = ["red", "orange", "yellow", "green", "blue", "indigo", "violet"]
for color in colors: 
    print(f"For rainbow, one of the colour is {color}")
```

```python
# using enumerate to get index and value of list items
fruits = ["apple", "banana", "cherry", "date", "strawberry"]
for i, fruit in enumerate(fruits):
    print(f"I like to eat {i} {fruit}")

    # Looping through a range of numbers  upto 10  
for i in range(1, 11):
    print(f"The current number is {i}")
```

Loop Control

```python
for i in range(1, 10):
    if i == 5:
        continue   # skip 5 number
    if i == 8:
        break        # stop when loop is eight
    print(i)
```

```python
# changes the elements in the list 
sports = ["cricket", "football", "tennis", "baseball", "volleyball" ]
for i in range(len(sports)):
    print("before", i, "is", sports[i])
    sports[i] = "good game"
    print("after", i, "is", sports[i])
```

## While Loop

```python
count = 5 
while count < 6:
    print(count)
    count += 1
print("loop finished")
```

```python
accuracy = 90
while accuracy <= 98:
    print(accuracy)
    accuracy += 1 
print(" accuracy achieved")
```

```python
image_processed = 5
while image_processed <= 10:
    print(f"processing the batch of {image_processed} ")
    image_processed += 2
```
