```python
# access through indexing
name = "Shahab Ahmad"
print(name[3])         
```

```python
# Slicing
name = "Shahab Ahmad"
print(name[0:5])
```

```python
# Converting to upper case
a = "Shahab Ahmad is beginner in python"
b = a.upper()
print(b)
```

```python
# finding and replacing something in the string
p = "Mary had a little lamb Little lamb, little lamb Mary had a little lamb \
Its fleece was white as snow And everywhere that Mary went Mary went, Mary went \
Everywhere that Mary went The lamb was sure to go"
print(p.find("snow"))
print(p.replace("Mary", "Shahab"))
print(p.split())
```

```python
# sub(), findall() function using regex module aka regular expression.
import re
str1 = " A quick brown fox jumps over the lazy dog"
new_str1 = re.sub(r"fox", "bear", str1) 
print(new_str1)

str2 = "How much wood would a woodchuck chuck, if a woodchuck could chuck wood?"
new_str2 = re.findall(r"woo", str2)
print(new_str2)
```
