# Data structures in Python
 
 ### Lists

```python
# Access through indexing 
A = ["Shahab Ahmad", 23, 3.2]
print(A[1])

# Slicing 
print(A[0:2])

#Adding elemnents using extend() and append()
B = ["Shahab Ahmad", 23, 3.2]
B.extend(["saidu sharif", 46])
print(B)

B = ["Shahab Ahmad", 23, 3.2]
B.append(["saidu sharif", 46])
print(B)

#List are mutable i.e we can change elements in it.
B = ["Shahab Ahmad", 23, 3.2]
B[0] = "Aizaz Ahmad"
print("After change", B)

# Deleting element from the list
C = ["Shahab Ahmad", 23, 3.2, "Abbas"]
del(C[3])
print("After Deleting", C)

#copy list
C = ["Shahab Ahmad", 23, 3.2, "Abbas"]     #changes in the one list will be reflected in other list too
E = C 
print("Copied", E)
C[0] = "Hazrat Usman"
print("Replace in the other list too", E)

 #Cloning in Lists
C = ["Shahab Ahmad", 23, 3.2, "Abbas"]    #changes will not be reflected in the D.
D = C [:]
print("Cloned", D)
```

### Tuples


All the above operation we do in the list are same as tuples, but there is one exception that tuples can't be
changed, they are immutable.


```python
music_types = ("pop", "rock", "soul", "hard rock", "soft rock","R&B", "progressive rock", "disco")

 # Find the length of Tuple
print(len(music_types))

#indexing of specific element
"progressive".find('v')

#sorting in tuple
tuple1 = ( 10, 9, 7, 8, 5, 6, 4, 2, 3, 1)
tuple2 = sorted(tuple1)
print(tuple2)
```

### Dictionaries

```python
soundtrack_dic = {"The Bodyguard":"1992", "Saturday Night Fever":"1977"}

# finding keys and values in the above dictionaries
soundtrack_dic.keys()                # dict_keys(['The Bodyguard', 'Saturday Night Fever'])
soundtrack_dic.values()              # dict_values(['1992', '1977'])


```

```python

# inventory scenario 

inventory = {}

ProductNo1 = { "item" : "Mobile Phone", "Quantity" : 5, "price" : 20000, "Release Year" : 2020}
inventory["ProductNo1"] = ProductNo1
ProductNo2 = { "item" : "tablet", "Quantity" : 10, "price" : 15000, "Release Year" : 2025}
inventory["ProductNo2"] = ProductNo2
print(inventory)

# checking if the items are there in the dictionary
  
 # print( "item" in inventory) it shows error, because the item is not top-level,
 # python only check top-level like product 1 and 2 in this case. 
print("item" in inventory["ProductNo1"])

# deletion 
del(inventory["ProductNo1"]["item"])
print(inventory)



```

### Sets

```python
#converting list to set using set()
list_mine = [ "Shahab Ahmad", 23, 3.2, "college colony","saidu sharif"]
set_mine = set(list_mine)
print(set_mine)

# Addition in set
set_mine.add("Abbas Ahmad")
print(set_mine)

#remove element from set 
set_mine.remove("Shahab Ahmad")
print(set_mine)

# intersection in sets
album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
intersection = album_set1 & album_set2
print(intersection)

# union in sets 
union = album_set1.union(album_set2)
print(union)

#difference in sets 
difference = album_set1.difference(album_set2)
print(difference)

# issubset
subset = album_set1.issubset(album_set2)
print(subset)
```
