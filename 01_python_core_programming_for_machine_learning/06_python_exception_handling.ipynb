### Exception Handling

```python
# A simple try and except block for handling exception.
try:
    result = 10/0
except ZeroDivisionError:
    print("you can not divide  by zero")
except:
    print("some other error encountered" )
```

```python
# value error
try: 
    num = int("xyz")
except ValueError: 
    print("the value you have entered can't be converterd into integer")
```

```python
a = 1 
try: 
    b = int(input("please enter a number to divide a"))
    a = a/b
    print("success a=", a)
except ZeroDivisionError:
    print("The number is o, can't be divided by 1")
except ValueError:
    print("please provide a valid value")
except:
    print("some went wrong, try again")
```
