## Conditions and Branching in python.

### If condition

```python
age = 18 

if age > 18:
    print("you can enter")

print("entry denied")

if age >= 18:
    print("you can enter")

print("you are not allowed") #Because of identation,it will run everytime, whether condition is t/f, becuse it
                              # doesn't belong the if condition. it is outside.
```

### Else

```python
age = 19 
if age < 18:
    print("you are underage, and not allowed")
else:
    print("you can enter")
```

### Elif

```python
age = 20 
if age > 20:
    print("you enter to the party")
elif age >= 20:
    print("go to the teens party")
else:
    print("move on")
```

### Logic operators

AND operator 


```python
birth_year = 2003
if birth_year > 2000 and birth_year < 2005:      #it will print the results, if both condition are true.
    print("eligble for master in abroad")
else:
    print("study here")
```

OR Operator

```python
birth_year = 2003
if birth_year > 2000 or birth_year > 2005:     #it will print the result, if one condition is true 
    print("eligble for master in abroad")       
else:
    print("study here")
```

NOT Operator

```python
birth_year = 2003
if not birth_year==2005:     
    print("not eligble for master in abroad")       
else:
    print("study here")
```
